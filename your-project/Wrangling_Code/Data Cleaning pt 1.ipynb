```python
import pandas as pd
import numpy as np
import re
import scipy.stats
```

```python
path = r'C:\Users\Usuario\Desktop\gits\Project-Week-5-Your-Own-Project\Data'
most_expensive = pd.read_csv(path+'/'+'most_expensive.csv')
```

```python
most_expensive.drop(['Unnamed: 0'], axis=1)
```

```python
string = most_expensive.ful_name[743]
re.findall(r'\w*\s|\w*-', string)
```

```python
exceptions =['New ','OFF-','OFF','Louis ','Louis','Fear','fear','a ','A ','dr.','Dr','Saint ','saint ']
```

```python
brand_list = []
for i in range(0,1000):
    string = most_expensive.ful_name[i]
    list_words = re.findall(r'\w*\s|\w*-', string)
    if list_words[0] in exceptions:
        name=list_words[0]+list_words[1]
        brand_list.append(name)

        
    else:
        name=list_words[0]
        brand_list.append(name)
    
```

```python
brand_list
```

```python
model_list = []
for i in range(0,1000):
    string = most_expensive.ful_name[i]
    list_words = re.findall(r'\w*\s|\w*-', string)
    if list_words[0] in exceptions:
        model=list_words[2:]
        model = ''.join(model)
        model_list.append(model)
    else:
        model=list_words[1:]
        model = ''.join(model)
        model_list.append(model)
        
   
```

```python
model_list
```

```python
string = most_expensive.avg_price_raw[998]
string
re.findall(r'(?<=Avg Sale\: \$)\d*,\d*|(?<=Avg Sale\: \$)\d*',string)
```

```python
price_list = []
for i in range(0,1000):
    string = most_expensive.avg_price_raw[i]
    number = re.findall(r'(?<=Avg Sale\: \$)\d*,\d*|(?<=Avg Sale\: \$)\d*',string)
    price_list.append(number)
```

```python
price_list = [price_list[i][0] for i in range(len(price_list))]
```

```python
price_list
```

```python
Most_Expensive_df =  pd.DataFrame({'brand':brand_list,
                     'model':model_list, 'price':price_list}, columns = ['brand','model','price'])
```
