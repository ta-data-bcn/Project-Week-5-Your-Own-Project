# Jordan 1 Chicago

```python
path = r'C:\Users\Usuario\Desktop\gits\Project-Week-5-Your-Own-Project\Data\raw'
airjordan1_Chi_df=pd.read_csv(path+'/'+'airjordan1_chicago.csv')
```

```python
#we open the df
airjordan1_Chi_df.drop(['Unnamed: 0'], axis = 1)
```

```python
#testing cell to write the function
string=airjordan1_Chi_df.prices[0]
number_raw = re.findall('(?<=\$)\d*,\d*|(?<=\$)\d*',string)
number_clean=number_raw[0].replace(',', '')
number_clean
```

```python
#defining the function that cleans the dollar sign and comma, check the int thing because it wasnt really working
def cleaner(x):
    for i in range(len(x)):
        string = x.prices[i]
        number_raw = re.findall('(?<=\$)\d*,\d*|(?<=\$)\d*',string)
        number_clean=number_raw[0].replace(',', '')
        x.prices[i]=int(number_clean)

    
```

```python
cleaner(airjordan1_Chi_df)
```

```python
airjordan1_Chi_df.drop(['Unnamed: 0'],axis = 1)
```

```python
#testing for the function
print(type(airjordan1_Chi_df.dates[4]))
date1 = datetime.datetime.strptime(airjordan1_Chi_df.dates[4], '%B %d, %Y')
type(datetime.datetime.strftime(date1, '%Y-%m-%d'))
```

```python
#defining the function that will turn the date column into an actual date object, then put it in the format we want
def date_parser(x):
    for i in range(len(x)):
        date_str = x.dates[i]
        date1 = datetime.datetime.strptime(date_str, '%B %d, %Y')
        date2 = datetime.datetime.strftime(date1, '%Y-%m-%d')
        x.dates[i] = date2
```

```python
date_parser(airjordan1_Chi_df)
```

```python
airjordan1_Chi_df.drop(['Unnamed: 0'],axis=1)
```

```python
#we change format to datetime again and  make a column for the weekday number
airjordan1_Chi_df['dates'] = pd.to_datetime(airjordan1_Chi_df['dates'],format = '%Y-%m-%d')
airjordan1_Chi_df['Num_day_week']=airjordan1_Chi_df['dates'].dt.weekday
```

```python
airjordan1_Chi_df.drop(['Unnamed: 0'],axis=1)
```

```python
#we make a column indicating the first day of the week for each day by computing the diference between the actual date and the days of the week.
airjordan1_Chi_df['First_day_week']=airjordan1_Chi_df.apply(lambda x:x['dates']-datetime.timedelta(days=x['Num_day_week']),axis=1)
```

```python
#checking the type for prices
airjordan1_Chi_df.drop(['Unnamed: 0'], axis = 1)
#airjordan1_Chi_df.groupby('First_day_week')['prices'].mean()
airjordan1_Chi_df['prices']=airjordan1_Chi_df.prices.astype(int)
airjordan1_Chi_df.dtypes
```

```python
#grouping by
avg_price_week=airjordan1_Chi_df.groupby('First_day_week')['prices'].mean()
```

```python
#turning the groupby into a df and saving as a csv because this is what we'll merge with the trends
avg_price_week_AJ1CHi=pd.DataFrame(avg_price_week)
avg_price_week_AJ1CHi.to_csv(r'C:\Users\Usuario\Desktop\gits\Project-Week-5-Your-Own-Project\Data\final'+'/'+'avg_price_week_AJ1CHi.csv')
```
