## Introduction

I've always been interested in demographic facts and trends, this project gave me the chance to explore one of the topics that I find more interesting, Earth's Overpopulation issue.
Ever since I studied Thomas R. Malthus's theory about the demographic trap ('Malthusian trap') shown below, I agreed completely with it and even thought I that I would live it some day. <img src="files/Malthusian_catastrophe.png" style="width: 250px;"/>

It states that the population grows exponentialy and the resources arithmetically, therefore, eventually, we would reach a point of catastrophe that could be solved in three ways. First, by natural causes such as natural catastrophes of pandemias such as the one we are living today. Second, by wars and conflicts big enough to impact the world population. Both of those scenarios picture a 'solution' where the outcome is that population drops, therefore resources are enough for the population. But the third option is an increase of productivity that allows us to produce enough recources than before and they can catch the population growth. 

Although when you look at it, the problem is that he never invisioned a scenario where we could stabilize population growth and, in doing so, avoid the malthusian trap for ever instead of just pushing it few years forward.
I can imagine that in his time, between the XVIII and XIX centuries, he couldn't imagine a future where a big part of the countries in Earth would have fertility rates below 2. But now, this is our reality, so I choose this topic to be able to dig into it, extract some conclusions and insights.

On one hand, This project is about understanding where earth's population growth occurs, meaning in which countries or continents, and which factors affect this growth the most. 
On the other hand, I want to use the data and my analysis to present a possible solution to this issue that is everybody's matter and that everyone should be worrying about it at some point.

#### About this topic I mainly had these three questions:

- Do less developed and less educated countries have a higher population growth than developed ones?
- What is the relationship between the GDP per capita and the population growth of each country? Does an increment of the first have anything to do with a decrease on the latter?
- Could it be that the solution to overpopulation lies on the developed world to help the undeveloped countries get to a confortable lifestyle so they stop growing?

### Without further delay, here are the results of my analysis

This is the headmap of correlation between the different parameters of the dataset.

From this map we can see two interesting relationships:
    - Strong positive correlation between Child Mortality and Fertility rates.
        This means that the higher the Child mortality of a country, 
        the more likely it is that its fertility rate is high as well.
    - Moderate negative correlation between GDP per capita and Fertility/Mortality rates. 
        This mean that the higer the GDP per capita of a country,
        the less likely it is that its Fertility and Child mortality rates are low.
        


This is the headmap of correlation between the different parameters of the dataset when we include the Tertiary education parameter.

From this map we can see two more interesting relationships:
    
    - Moderate positive correlation between Education and and GDP per capita.
        This means that the more educated a country is, 
        the more likely it is that its GDP per capita is high as well.

    - Moderate negative correlation between Education and both Fertility/Mortality rates. 
        This means that the more educated a country is,
        the more likely it is that its Fertility and Child mortality rates are low.
       

### I know you might want to see those relationships in a more graphic way, here it is:

This shows the negative correlation between Fertility and Child Mortality rates. This plot is from 1981 and we will see below how it's progressed in 2015.

From these two charts, we can see that:
- Most countries move from high fertility and high mortality towards the bottom left.
- Most of the world is shifting to a growth stabilization  scenario.
- In 2015 still many countries are on the top left. Mainly Africa and the undeveloped countries. 


This shows the negative correlation between the Fertility and the GDP per capita. This plot is from 1981 and we will see below how it's progressed in 2015.

From these two charts, we can see that:

- Most countries move from high fertility and low GDP per capita towards the   bottom right.
- This shows us that the higher the GDP per capita, the lower the Fertility rate.
- In 2015 most countries stopped growing due to an increase of their ‘Standards of living’. But there's a long road ahead before Africa and the undeveloped countries can reach this point.


## Conclusions

- World Population grows mainly in the undeveloped world.

- Education and GDP per Capita have a negative impact on the Fertility rate. In other words, the more education and quality of   life, the less likely is for a woman to have kids.

- According to our data, one possible solution to overpopulation Is closing the economic and educational gaps between the        developed and undeveloped world.
