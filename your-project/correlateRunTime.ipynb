```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
# It is time now to get more data from Netflix of tv shows and movies!!!
# reutruns pandas series if there is only one colunmn 
imdb_mov = pd.read_csv('../dataset/IMDb_movies.csv', usecols = ['imdb_title_id','genre','duration','country','avg_vote','votes','reviews_from_users'])
```

```python
idx  = imdb_mov.country == 'USA'
idx2 = imdb_mov.votes > 100000
idx3 = (imdb_mov.avg_vote >=6) & (imdb_mov.avg_vote <=8)
imdb_mov[idx & idx2 & idx3].avg_vote.hist(bins=20)
```

```python
idx4 = (imdb_mov.duration >=70) & (imdb_mov.duration <=180)
imdb_mov[idx & idx2 & idx3 & idx4].duration.hist(bins=30)
```

```python
fig = plt.figure(figsize=[7,5],constrained_layout=True)
fig.suptitle('Avg_vote vs Duration - France', size=16, weight='bold',y=1.05)
spec = gridspec.GridSpec(ncols=1, nrows=1, figure=fig)
label_prop = {'size':14,'family':'fantasy','style':'oblique'}
axis = fig.add_subplot(spec[0,0])
plot_var = imdb_mov[idx&idx2&idx3&idx4].plot.scatter('duration','avg_vote',ax=axis,s=20,alpha=0.6,marker='o',ec='silver',fc='mediumblue')
plt.sca(plot_var)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('Duration',fontdict=label_prop)
plt.ylabel('Avg_vote',fontdict=label_prop)
# plt.xlim(0,80)
# plt.ylim(0,550)
# plt.grid(which='major', axis='y')
```
