### Comparison and relation between popularity and price

The most popular collaborator on average is Michael Jordan an the most expensive sneaker is the Air Yeezy 2 Red October, both of them also have the highest standard deviations. These two are also the two with the highest correlation between popularity and price.

So we can affirm that Michael Jordan's popularity has the most effect on the price of his signature sneaker, and a positive one, in comparison to that of other younger, newer figures, even though the pricing of his sneaker on reselling is the lowest among them.

The second most correlated celebrity is Kanye West, this time negatively, which would make sense given the fact that tipically the headlines he makes are quite negative, which would mean people have less of an incentive to be associated with him and prices would fall down because of a demand drop.

Both Travis and Virgil's popularity seem to have less of an impact on the pricing of their sneaker on the resell market. However there's the caveat that there are considerations outside the scope of this paper that might lead to this watered down relation, such as the relationship not being linear or there being a lag between the searches and it's  reflection on the price. There is also the fact that there are other, probably stronger variables that would explain the changes in price, and future revisions of this project will touch on topics like, size, availability and media exposure of the product (instagram posts, youtube videos, product placement, etc...)

- Kanye

For Kanye the relationship is a tad less obvious and visually represented it seems more probable that there is a lag between popularity and prices, suggesting that the controversies might come before the price change. The second peak illustrates his meeting with Trump in the White House. However none of these events significantly affects the price.

The price tremendously soars on March of this year when Kanye stole the show with his Yeezy colection at the Paris Fashion Week, further confirming his status in the fashion industry.
