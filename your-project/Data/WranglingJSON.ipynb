```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Gen Stats For 2019
MLS2019_GenStats = pd.read_json("2019_GenStats.jscsrc")
MLS2019_GenStats
```

```python
# 2019 standings eastern conference
MLS2019_east_standings = pd.read_json("2019_east_standings.jscsrc")
MLS2019_east_standings
```

```python
# Merging the two dataframes
MLS2019 = MLS2019_east_standings.merge(MLS2019_GenStats)
MLS2019['Season'] = '2019'
MLS2019
```

```python
## Linnear regression test

slope, intercept, r_value, p_value, std_err = stats.linregress(MLS2019['G+A-PK'], MLS2019['W'])
print ('The slope is: ' + str(slope))
print ('The intercept is: ' + str(intercept))
print(p_value)
print ('The correlation is: ' + str(MLS2019['G+A-PK'].corr(MLS2019['W'])))
print ('The covarience is: ' + str(MLS2019['G+A-PK'].cov(MLS2019['W'])))
sns.regplot(x='G+A-PK', y='W' , data=MLS2019)
plt.show()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(MLS2019['G+A-PK'], MLS2019['Pts'])
print ('The slope is: ' + str(slope))
print ('The intercept is: ' + str(intercept))
print(p_value)
print ('The correlation is: ' + str(MLS2019['G+A-PK'].corr(MLS2019['Pts'])))
print ('The covarience is: ' + str(MLS2019['G+A-PK'].cov(MLS2019['Pts'])))
sns.regplot(x='G+A-PK', y='Pts' , data=MLS2019)
plt.title('Linear Regression Plot')
# Set x-axis label
plt.xlabel('Goals+Assists-PK')
# Set y-axis label
plt.ylabel('Season Points')
plt.figure(figsize = (24, 10))
plt.show()
```

# Attacking makes you win matches, but defense wins championships

This is a common mantra between coaches of team sports. I would like to know what's the real influence of Defensive game into winning a championship in the long run. I would measure this by the correlation between defensive KPI and the total points in a season. Analysis is performed over 2018 MLS Champions Atlanta FC. i want to check for the effectiveness of Gerardo Martino's deffensive tactics of high pressure and tackling contundence.

Specifically i want to test two hypothesis:

### 1. Defense effectiveness gives you more points
We will check the effectivity against its rivals in the Eastern Conference and also how this effectiveness affects the team capacity to generate points. GA (Goals against) vs Points

### 2. Gerardo Martino's high pressure defense style generates more goals
We will compare the different teams attacking 3rd pressure and attacking effectiveness correlated with it's capacity of generating more GF (Goals for)

```python
labels=['Press', 'Succ', 'Tkl', 'TklW', 'Int']
markers = [0, 1, 2, 3, 4, 5]
str_markers = ["0", "1", "2", "3", "4", "5"]

def make_radar_chart(name, stats, attribute_labels = labels, plot_markers = markers, plot_str_markers = str_markers):

    labels = np.array(attribute_labels)

    angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)
    stats = np.concatenate((stats,[stats[0]]))
    angles = np.concatenate((angles,[angles[0]]))

    fig= plt.figure()
    ax = fig.add_subplot(111, polar=True)
    ax.plot(angles, stats, 'o-', linewidth=2)
    ax.fill(angles, stats, alpha=0.25)
    ax.set_thetagrids(angles * 180/np.pi, labels)
    plt.yticks(markers)
    ax.set_title(name)
    ax.grid(True)

    fig.savefig("SpiderChart.png" % name)

    return plt.show()

make_radar_chart("MLS_Def_East", [2,3,4,4,5]) # example
```
