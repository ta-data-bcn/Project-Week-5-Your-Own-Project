# Paper of Digging into Pronto (Cycle Share Company from Seattle(US))

The purpose of this paper is to describe step by step what has been done from the beginning until the end of the project.

## Introduction

### About the project, description, motivation and questions

The Pronto Cycle Share system consists of 500 bikes and 54 stations located in Seattle. The Cycle Share company initialized its activities on 2014 but only survived 3 years, on 2017 the company close the service because they did not obtained profits.

Reading about what happened with Pronto Cycle Share, a question appear in my mind, it was "WHY?". Why, Pronto a cycle share company was dismantled in three, which where the reasons? Maybe the weather during these years affected? Was only a problem of profits for the company it means less users year by year? Too many stations for a few people?

Due my interest on the sustanaible mobility in the city and my interest on bycicles I decided choose this topic and started to find interesting databases.

### Databases

The databases used for the projects has been two. One with the information related to the trips, durations, type of members and so on. This has been obtained from the Mayor Govertment Seattle website. 
The second database was related to the weather in Seattle from 1984 to 2018.

## Process explained

### Data Cleaning
First of all I started with the data cleaning of the database after reading the .csv file.

#### Database Trips and Members
The data cleaning consisted on checking how many rows and what is the appearance of the dataframe. After that I noticed the "date" column had a not proper type of data, it was object instaed datetime type, this has been changed and a new was created to save the results on it.
A table was merged in order to add therefore this table to the dataframe weather. This merged table is compounded with the date, the sum of the trips duration per day and the number of the trips per day as well.

#### Database Weather
Exactly the same process as the database before, changed the type of the date and implemented a for loop to modify the number of the months in order to begin with a 00,01,02,03...
The dataframe merged before now is used to add next to the weather dataframe.

### Visualization
The second step has been the analysis of the data once has been cleaned. The results has been showed in a form of graph using the matplot library to represent them.

### Results

In this first results is shown the use in hours over the days that the service was available. Using two SMA (Simple Moving Average) one of the 30 periods and the other of 120 periods, we cant see how is the main trend during this time, is obvious that the decreasing maximums along the days show a negative trend meaning the use is descreasing in general terms year by year. No goods results.

The second graph shows the relation between the two types of membership possible. Member is a monthly membership while Short-Term Pass Holder is a membership for a 24 hours. We can appreciate the difference of relative decreasing memberships, from 2015 to 2016 the in relative ways the Member type was dramatically decreased while the Short-Term didn't it represents a not healthy business due any business need a fix income to survive and following this trend it seems not to happen.

The last graph we can see the relation with weather (rainy days) and number of rides during the months of the service.
There is an obviusness inverse proportion of the number of rides during the rainy days, this could have been another reason to add for the desaparition of the Cycle Share company, Pronto.
