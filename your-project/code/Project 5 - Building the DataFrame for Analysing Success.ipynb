## Success by not being closed

```python
# making dataframe for counting growth rate
comps_closed_ann = comps_acq_joined_time.groupby(['year_closed']).count()
comps_closed_ann = comps_closed_ann.reset_index()
comps_closed_ann = comps_closed_ann[['year_closed', 'year_founded', 'name']]
comps_closed_ann.rename(columns = {'name' : 'number'}, inplace = True)
```

```python
idx = comps_closed_ann[comps_closed_ann['year_closed'] < 1998].index
comps_closed_ann.drop(idx , inplace=True)
comps_closed_ann.year_closed.min()
```

```python
comps_closed_ann.head()
```

```python
comps_closed_ann['yoy_closed'] = (comps_closed_ann['number'] -comps_closed_ann['number'].shift(1)) / comps_closed_ann['number'].shift(1)
comps_closed_ann.tail()
```

```python
idx = comps_closed_ann[comps_closed_ann['year_closed'] >= 2013].index
comps_closed_ann.drop(idx , inplace=True)
comps_closed_ann.year_closed.max()
```

```python
companies_industry_year = pd.read_csv(r'data/companies_industry_year.csv')
companies_industry_year.head()
```

```python
comps_founded_closed = companies_industry_year.copy()
idx = comps_founded_closed[comps_founded_closed['year_founded'] < 1999].index
comps_founded_closed.drop(idx , inplace=True)
comps_founded_closed.year_founded.min()
```

```python
comps_founded_closed.head()
```

```python
comps_founded_closed = comps_founded_closed.drop(columns = 'country_code')
```

```python
comps_founded_closed = comps_founded_closed.groupby(['category_code', 'year_founded']).sum()
comps_founded_closed = comps_founded_closed.reset_index()
comps_founded_closed.head()
```

```python
comps_founded_closed['ratio_fc'] = comps_founded_closed['closed_at'] / comps_founded_closed['founded_at']
comps_founded_closed.head()
```

```python
comps_founded_closed_2 = comps_founded_closed[['category_code', 'ratio_fc', 'id']]
```

```python
comps_founded_closed_2 = comps_founded_closed_2.groupby(['category_code']).agg({'ratio_fc':'mean','id':'sum'})
comps_founded_closed_2 = comps_founded_closed_2.reset_index()
comps_founded_closed_2.head()
```

```python
# comps_founded_closed_2.to_csv(r'data/comps_ratio_fc.csv')
```

```python
# companies_yig = pd.read_csv(r'data/companies_year_industry_geography.csv')
# companies_yig.head()
```

Correlation

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
sns.set(style="white", context="talk")
fig, ax = plt.subplots(figsize=(16,10)) 
x = comps_founded_closed_2['ratio_fc']
y = comps_founded_closed_2['id']
ax = sns.regplot(x, y, 
                  data = comps_founded_closed_2, scatter_kws = {"s": 250},
                  marker = "+", color = 'r')
ax.set(xlabel = "Ratio of Companies closed in Industry (%)", ylabel = "Number of Companies in the Industry")
result = stats.linregress(x, y)
print("Slope: ", result.slope)
print("Intercept: ", result.intercept)
print("rvalue: ", result.rvalue)
print("pvalue: ", result.pvalue)
print("stderr: ", result.stderr)
```
