```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
```

```python
fname = 'vicky_table.html'
HtmlFile = open(fname, 'r', encoding='utf-8')
source_code = HtmlFile.read() 
```

```python
soup = BeautifulSoup(source_code, 'html.parser')
```

```python
rows = soup.find_all('tr')
columns = rows[0].getText().strip().split('\n')

list_of_rows = []
list_to_clean = []

for row in rows:
    list_of_cell = []
    company = row.getText().strip('\t').split('\n')
    for i in company:
        i = re.sub('\t', '', i)
        i = re.sub(' ', '', i)
        i = re.sub('  ', '', i)
        if i:
            list_of_cell.append(i)
    if len(list_of_cell) == 6:
        list_of_rows.append(list_of_cell)
    elif len(list_of_cell) > 6:
        list_to_clean.append(list_of_cell)
    
del list_of_rows[0]
```
