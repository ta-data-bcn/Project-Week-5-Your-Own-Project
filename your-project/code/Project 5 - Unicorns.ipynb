## Unicorns

```python
import pandas as pd
import numpy as np
```

```python
unicorns = pd.read_csv(r'data/cbinsights_entire_unicorn_tracker_sorted.csv').drop(columns = 'Unnamed: 0')
unicorns.head(50)         
```

```python
# By Industry / Average Valuation
```

```python
unicorns.Industry.unique()
```

```python
unicorns_by_industry = unicorns.copy()
unicorns_by_industry = unicorns_by_industry.groupby(['Industry']).agg({'Company':'count','Valuation ($B)':'mean'})
unicorns_by_industry = unicorns_by_industry.reset_index()
```

```python
unicorns_by_industry.rename(columns = {'Company' : 'Number'} , inplace = True)
unicorns_by_industry['Percent of Total'] = unicorns_by_industry['Number'] / len(unicorns.Company)
unicorns_by_industry = unicorns_by_industry.sort_values(by = "Percent of Total", ascending = False).reset_index(drop=True)
unicorns_by_industry.head()
```

```python
# unicorns_by_industry.to_csv(r'unicorns_by_industry.csv')
```

```python
# By Country / Average Valuation
```

```python
unicorns_by_country = unicorns.copy()
unicorns_by_country = unicorns_by_country.groupby(['Country']).agg({'Company':'count','Valuation ($B)':'mean'})
unicorns_by_country = unicorns_by_country.reset_index()
unicorns_by_country.rename(columns = {'Company' : 'Number'} , inplace = True)
unicorns_by_country['Percent of Total'] = unicorns_by_country['Number'] / len(unicorns.Company)
unicorns_by_country = unicorns_by_country.sort_values(by = "Percent of Total", ascending = False).reset_index(drop=True)
unicorns_by_country.head()
```
