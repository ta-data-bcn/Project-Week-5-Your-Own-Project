```python
#here we import the needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datapackage
import re
```

## Population from 1981 to 2018

```python
Pop_df = pd.read_excel('HistoricalPopulationValues.xls')
Pop_df
```

## GDP in billions($)

```python
Gdp_df = pd.read_excel('GDP_billions.xls')
Gdp_df
```

```python
Gdp_df = Gdp_df.drop([1980], axis=1)
```

## GDP per capita

```python
columns = list(Gdp_df.columns)
columns.remove('Country')
gdp_melted = pd.melt(Gdp_df, id_vars=['Country'], value_vars=columns, var_name="Year", value_name="GDP")

gdp_melted.head()
```

```python
columns = list(Pop_df.columns)
columns.remove('Country')
pop_melted = pd.melt(Pop_df, id_vars=['Country'], value_vars=columns, var_name="Year", value_name="Population")

pop_melted.head()
```

```python
gdp_p_df = pd.merge(gdp_melted, pop_melted, on=['Country','Year'])
gdp_p_df.head()
```

```python
gdp_p_df['GDP_per_capita'] = ((gdp_p_df['GDP'] * 1000000000) / gdp_p_df['Population'])
```

```python
gdp_p_df.head()
```

```python
gdp_p_df.to_csv('GDP_P.csv')
```

```python
gdp_p_years = gdp_p_df[['Year','GDP','Population', 'GDP_per_capita']].groupby(['Year']).agg({'GDP':'mean','Population':'mean', 'GDP_per_capita':'mean'})
gdp_p_years.head()
```

```python
gdp_p_years['GDP_in_$'] = gdp_p_years['GDP'] * 1000000000
```

```python
gdp_p_years.head()
```

```python
gdp_p_years.plot()
```

```python
plt.scatter(gdp_p_years['Population'], gdp_p_years['GDP_in_$'])
```

## Fertility  AND CHILD MORTALITY
