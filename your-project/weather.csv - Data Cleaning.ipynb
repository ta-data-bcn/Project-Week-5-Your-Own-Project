```python
df_lol["YearMonthDay"] = df_lol["YearMonthDay"].astype("datetime64[ns]")
```

```python
df_general = df_lol.merge(df_weather, left_on="YearMonthDay", right_on="Date")
```

```python
df_general = df_general.drop(["Date"], axis=1)
```

```python
df_general = df_general.rename(columns={"usertype":"rides"})
```

```python
df_general.head()
```

```python
df_general.index = df_general['YearMonthDay']
```

```python
x_list_months = sorted(df_general.index.month.unique())
```

```python
df_general["Month"] = df_general["YearMonthDay"].dt.month
```

```python
def replace_numbers(month):
    if month < 10:
        return "0" + str(month)
    return str(month)

df_general["Month"] = df_general["Month"].apply(replace_numbers)
```

```python
df_general["Year"] = df_general["YearMonthDay"].dt.year
```

```python
df_general["YearMonth"] = df_general[["Year","Month"]].apply(lambda row: str(row["Year"])+"-"+str(row["Month"]), axis=1)
```

```python
df_wololo = df_general.groupby(["YearMonth"]).sum()[["Precipitation_In","rides"]]
df_wololo.head()
```

```python
fig, ax1 = plt.subplots(1, figsize=(25, 15))

ax2 = ax1.twinx()

ax1.fill_between(df_wololo.index, df_wololo["Precipitation_In"], alpha=0.4)
ax1.set_title("Rides done related to precipitation over months", fontdict={"fontsize":25}, pad=20)
ax1.set_ylabel("Precipitation [Inches]", fontdict={"fontsize":20}, labelpad=20)
ax2.set_ylabel("Number of rides", fontdict={"fontsize":20},  labelpad=20)
ax1.set_xlabel("Month", fontdict={"fontsize":20}, labelpad=20)
ax2.plot(df_wololo.index,df_wololo["rides"], color="purple", linestyle="--", marker="o", alpha=1, label="rides")
ax1.set_xticklabels(df_wololo.index, fontsize=20, rotation=45)
ylabels = ax1.get_yticks()
ax1.set_yticklabels(ylabels, fontsize=20)
ylabels = ax2.get_yticks()
ax2.set_yticklabels(ylabels, fontsize=20)
plt.legend(fontsize=20)
plt.show()
```
