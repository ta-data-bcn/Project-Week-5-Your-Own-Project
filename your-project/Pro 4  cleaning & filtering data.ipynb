## GOAL :

- Find good jobs by salary, company rating and locations


```python
import pandas as pd
import numpy as np
```

### Data exploration

```python
df = pd.read_csv('DataAnalyst.csv', index_col=0)
```

```python
df
```

### Change value in Easy Apply  -1 to False

```python
df['Easy Apply'].value_counts()
```

```python
df['Easy Apply'].replace({"-1": "False"}, inplace=True)
```

```python
df
```

### Replace missing value

As some of the columns contains -1 or '-1.0' or '-1' etc . We need to clean this(This is kind of null values)

```python
df.replace(to_replace =-1 , value=np.nan,inplace=True)
df.replace(to_replace ='-1' , value=np.nan,inplace=True)
df.replace(to_replace =-1.0 , value=np.nan,inplace=True)
```

```python
df
```

### Check for missing value: 

```python
df.isnull().sum()
```

#### Drop Competitors and Founded columns because the values are not useful 

```python
df.drop(['Competitors'], axis = 1, inplace = True)
```

```python
df.drop(['Founded'], axis = 1 , inplace = True)
```

```python
df 
```

### Split salary into two columns min salary and max salary.


```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace('(Glassdoor est.)' ,'')
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace(' ( )' ,'')
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace('$' ,'')
```

```python
df['Salary Estimate'] = df['Salary Estimate'].str.replace('K' ,'')

df['Salary Estimate']
```

```python
df[['Min Salary','Max Salary']] = df['Salary Estimate'].str.split("-",expand=True,)
```

### Clear company name

```python
df['Company Name']
```

```python
df['Company Name']=df['Company Name'].str.split('\n', 1)
```

```python
df['Company Name']
```

### Filtering good companiesÂ´ s ratings are higher than 4


```python
df['Rating'].value_counts().sort_values(ascending=False)
```

```python
good_company = df.loc[df['Rating'] >= 4]
```

### Filtering companies with good salaries

```python
good_company = good_company[good_company['Salary Estimate'].notna()]
```

```python
good_company.isnull().sum() 
```

#### Find mean of min salary

```python
min_salary= good_company['Min Salary'].tolist()

min_salary = np.array(min_salary)

min_salary1=[] 

for element in min_salary:
    element.replace('.' ,'')
    min_salary1.append(element)
min_salary1 

```

```python
min_salary2 = []
for i in min_salary1 :
    i = int(i)
    min_salary2.append(i)
min_salary2    
```

```python
mean_min_salary = np.mean(min_salary2)
mean_min_salary 
```

```python
good_company['Min Salary'] = min_salary2   
```

#### Good companies salaries are higher than mean of min salary which is 55k

```python
good_salary = good_company[good_company['Min Salary'] >= 55]
```

```python
good_salary
```

### Find good locations 

```python
good_salary['Location'].unique()
```

#### Good locations based in  https://www.getkisi.com/work-life-balance-usa#table

```python
good_jobs = good_salary.loc[(good_salary.Location == "San Diego, CA") | (good_salary.Location == "New York, NY")| (good_salary.Location == "San Francisco, CA")| (good_salary.Location == "Seattle, WA")| (good_salary.Location == "Los Angeles, CA")| (good_salary.Location == "Denver, CO")]
```

```python
good_jobs['Location'].value_counts()
```

```python
good_jobs = pd.DataFrame(data= good_jobs)
```

```python
good_jobs.to_csv("good_job2.csv", sep=";")
```
